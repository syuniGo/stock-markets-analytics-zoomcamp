```python
import inspect

def calRange(df):
  frame = inspect.currentframe()
  print(len(inspect.getargvalues(frame)))
  args, _,_, values = inspect.getargvalues(frame)
  # print('function name "%s"' % inspect.getframeinfo(frame)[0])
  # for i in args:
  #   print('    %s = %s' % (i, values[i]))
  max =df['Close'].max()
  print(max)
  min =df['Close'].min()
  res = (max - min )/max
  print(res)
  re = math.ceil(res*100*100)/100
  print(f'+{re}%')
  return re
```

```python
def caldividends(df):
  stock = yf.Ticker(df)
  hist = stock.history(start="2023-01-01",end="2023-12-31")
  # print(hist.info)
  sum = hist['Dividends'].sum()
  last_p = hist.iloc[-1]['Close']
  dividendYield = sum/last_p
  re = math.ceil(dividendYield*100*10)/10

  print('re',re)
  return re
```

```python
caldividends('2222.SR')
caldividends('BRK-B')
caldividends('AAPL')
caldividends('MSFT')
caldividends('GOOG')
caldividends('JPM')
```
